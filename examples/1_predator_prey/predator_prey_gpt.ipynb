### Creating your cadcadGPT agent

```python
from cadcad_gpt import CadCAD_GPT
```

Pass it your openai_key, and radcad model, simulation and experiment objects. Additionally you can pass your model docs as well.

```python
cadcad_gpt = CadCAD_GPT(openai_key, model, simulation, experiment, docs)
```

We can now ask questions in natural language to our cadcad_gpt chatbot. The Planner and Executor Agents will solve the user questions using tools available to them in the toolkit.

```python
cadcad_gpt('what is the model about?')
```

```python
cadcad_gpt('How is the prey population calculated?')
```

```python
cadcad_gpt('what are params of the model?')
```

```python
cadcad_gpt('can you plot predator and prey populations?')
```

```python
cadcad_gpt('can you plot the prey population against predator population as a phase diagram with timestep as color?')
```

```python
cadcad_gpt('what is the avg predator population?')
```

```python
cadcad_gpt('what are the descriptive stats of the population of the prey?')
```

### Multi hop Questions

```python
cadcad_gpt('what is my current prey reproduction rate, change it to triple its current value, plot predator and prey populations as a line chart?')
```

```python
cadcad_gpt('can you plot the prey population and predator population as a line chart?')
```

We can chain a bunch of tasks together

```python
cadcad_gpt('whats the current prey reproduction rate, change it to 2, find out the avg value of predator column, plot the predator population column?')
```

```python
cadcad_gpt('plot the prey population column, change the prey reproduction rate to 0.7 and then plot the prey population column again ?')
```
